```
from victor_hub.skills.nlp_skill import NLPSkill

nlp = NLPSkill()
print("‚úÖ NLP Skill loaded")
print(f"\nCapabilities: {', '.join(nlp.capabilities)}")
```

## üè∑Ô∏è Named Entity Recognition

```
text = "Victor is an AGI system developed by MASSIVEMAGNETICS in 2025."

result = nlp.execute({"task": "entities", "text": text})

print(f"Text: {text}\n")
for ent in result['entities']:
    print(f"  {ent['text']:20s} ‚Üí {ent['label']}")
```

## üîë Keyword Extraction

```
text = """Quantum-fractal cognition combines quantum interference 
with fractal geometry for advanced AI reasoning."""

result = nlp.execute({"task": "keywords", "text": text, "top_n": 5})

print("Keywords:")
for kw in result['keywords']:
    print(f"  ‚Ä¢ {kw}")
```

## üìä Part-of-Speech Tagging

```
text = "Victor processes quantum patterns efficiently."

result = nlp.execute({"task": "pos_tags", "text": text})

for token in result['pos_tags']:
    print(f"  {token['text']:15s} {token['pos']:6s} {token['tag']}")
```
